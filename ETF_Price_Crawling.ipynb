```python
url = 'https://finance.naver.com/item/sise_day.nhn?code=139260'
headers = {'user-agent' : UserAgent().chrome} #UserAgent 숨기기
response = requests.get(url, headers = headers)
# doc = BeautifulSoup(response.text, 'html.parser') #이거 안됨
doc = BeautifulSoup(response.text, 'html5lib') #속도가 느림 / 정확 
doc.select
```
